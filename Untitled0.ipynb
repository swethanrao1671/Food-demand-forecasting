```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv('/content/drive/MyDrive/dataset/train.csv')

# Print the first 10 rows of the data
print(df.head(10))

# Print summary statistics of the data
print(df.describe())

# Create a histogram of the number of orders
plt.hist(df['num_orders'], bins=20)
plt.xlabel('Number of orders')
plt.ylabel('Frequency')
plt.show()

# Create a line chart of the number of orders over time
orders_by_week = df.groupby('week')['num_orders'].sum()
plt.plot(orders_by_week.index, orders_by_week.values)
plt.xlabel('Week')
plt.ylabel('Number of orders')
plt.show()

# Create a bar chart of the average number of orders by center
orders_by_center = df.groupby('center_id')['num_orders'].mean()
plt.bar(orders_by_center.index, orders_by_center.values)
plt.xlabel('Center ID')
plt.ylabel('Average number of orders')
plt.show()
```
